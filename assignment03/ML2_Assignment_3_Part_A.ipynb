**Active learning**

Η ενεργή μάθηση (active learning) είναι μια μέθοδος μηχανικής μάθησης όπου το μοντέλο εκπαιδεύεται να ζητάει ετικέτες για συγκεκριμένα δείγματα που θεωρεί ότι θα επιφέρουν τη μεγαλύτερη βελτίωση στην απόδοσή του. Σε αντίθεση με την παραδοσιακή μέθοδο μηχανικής μάθησης που απαιτεί ένα πλήρες σύνολο ετικετών για την εκπαίδευση, το ενεργό μοντέλο μπορεί να ζητήσει νέες ετικέτες για συγκεκριμένα δείγματα κατά τη διάρκεια της εκπαίδευσης.

Συνήθως, η επιλογή των δειγμάτων για την οποία ζητείται η ετικέτα γίνεται με βάση την αβεβαιότητα του μοντέλου σχετικά με αυτά τα δείγματα. Πιο συγκεκριμένα, τα δείγματα που προκαλούν αβεβαιότητα στο μοντέλο, όπως τα δείγματα που ανήκουν σε ασαφείς κατηγορίες ή έχουν κοντινές προβλέψεις, θεωρούνται κρίσιμα και απαιτούν ετικετοποίηση από ανθρώπινους εμπειρογνώμονες.

Η ενεργή μάθηση είναι χρήσιμη όταν τα δείγματα είναι δαπανηρά ή δύσκολα να αποκτηθούν ετικέτες και μπορεί να μειώσει τον αριθμό των δειγμάτων που απαιτούνται για την εκπαίδευση ενός μοντέλου μηχανικής μάθησης, επιτρέποντας την απόκτηση υψηλής απόδοσης με μικρότερο κόστος.

**Query-by-Committee**

Το Query-by-Committee είναι μια μέθοδος active learning που βασίζεται σε μια επιτροπή (committee) από εκτιμητές (estimators) για να καθορίσει ποια δείγματα πρέπει να επιλεγούν για ετικετοποίηση. Η επιτροπή αποτελείται από πολλαπλούς εκτιμητές, ο καθένας από τους οποίους εκπαιδεύεται σε ένα υποσύνολο των δεδομένων εκπαίδευσης.

Κατά τη διάρκεια του active learning, η επιτροπή χρησιμοποιείται για να προβλέψει τις ετικέτες των αδιευκρίνιστων δειγμάτων στον χώρο επιλογής (pool set). Οι εκτιμητές στην επιτροπή συνεργάζονται για να παράγουν προβλέψεις και μετρούν την ασυμφωνία (disagreement) μεταξύ των προβλέψεών τους για κάθε δείγμα. Η ασυμφωνία μπορεί να υπολογιστεί με βάση τη διακύμανση ή άλλες μετρικές.

Στη συνέχεια, το δείγμα που έχει τη μεγαλύτερη ασυμφωνία μεταξύ των εκτιμητών επιλέγεται για ετικετοποίηση και προστίθεται στο σύνολο εκπαίδευσης. Αυτή η διαδικασία επαναλαμβάνεται για έναν προκαθορισμένο αριθμό επαναλήψεων ή μέχρι να εκπνεύσει ο χρόνος εκτέλεσης.

Ο αλγόριθμος Query-by-Committee αξιοποιεί την ποικιλία των εκτιμητών για να επιλέξει δείγματα που είναι πιο αβέβαια ή δύσκολα για την επιτροπή να αποφασίσει. Αυτό του επιτρέπει να εστιάσει σε περιοχές του χώρου όπου υπάρχει αβεβαιότητα και να βελτιώσει την απόδοση του μοντέλου με λιγότερες ετικέτες.



**Το "committee size" αναφέρεται στον αριθμό των εκτιμητών που απαρτίζουν την επιτροπή.**

**Query-by-Committee & Uncertainty Sampling**






H τεχνική "Query-by-Committee" (QBC) συνδυάζεται συχνά με τη μέθοδο "Uncertainty Sampling" στην ενεργή μάθηση. Η QBC χρησιμοποιεί μια επιτροπή από μοντέλα μηχανικής μάθησης για να αξιολογήσει την αβεβαιότητα και να πάρει αποφάσεις για την επιλογή δειγμάτων που πρέπει να ετικετοποιηθούν.

Συγκεκριμένα, η QBC λειτουργεί ως εξής: Η επιτροπή αποτελείται από πολλαπλά μοντέλα μηχανικής μάθησης που έχουν εκπαιδευτεί σε ένα αρχικό σύνολο δεδομένων. Στη συνέχεια, για κάθε ανεπισημασμένο δείγμα, κάθε μοντέλο της επιτροπής κάνει μια πρόβλεψη. Η αβεβαιότητα υπολογίζεται με βάση τις ποικίλες προβλέψεις των μοντέλων, όπως η διακύμανση των προβλέψεων ή ο αριθμός των αντικρουόμενων προβλέψεων.

Στη συνέχεια, τα δείγματα με τη μεγαλύτερη αβεβαιότητα επιλέγονται για ετικετοποίηση από ανθρώπινους εμπειρογνώμονες. Οι ετικέτες των δειγμάτων προστίθενται στο αρχικό σύνολο δεδομένων και το μοντέλο εκπαιδεύεται ξανά. Αυτή η διαδικασία επαναλαμβάνεται για πολλούς κύκλους, ενισχύοντας σταδιακά την απόδοση του μοντέλου.

Ο συνδυασμός της QBC με την uncertainty sampling επιτρέπει τη χρήση της αβεβαιότητας ως κριτήριο επιλογής για την ετικετοποίηση δειγμάτων. Με αυτόν τον τρόπο, το μοντέλο μπορεί να εστιάσει σε δείγματα που παρουσιάζουν μεγαλύτερη αβεβαιότητα και να βελτιώσει την απόδοσή του με λιγότερους ανθρώπινες πόρους.

**Active Learning with Uncertainty sampling & Random sampling:**



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from modAL.models import ActiveLearner
from modAL.uncertainty import uncertainty_sampling
from modAL.models import CommitteeRegressor

# Set our RNG seed for reproducibility.
RANDOM_STATE_SEED = 123
np.random.seed(RANDOM_STATE_SEED)

def random_sampling(X, n_samples):
    n_pool = len(X)
    if n_samples > n_pool:
        raise ValueError("Pool set size is smaller than the requested number of samples.")
    query_indices = np.random.choice(range(n_pool), size=n_samples, replace=False)
    return query_indices

# Load the diabetes dataset
diabetes = load_diabetes()
X_raw = diabetes.data
y_raw = diabetes.target

# Split the test set into an unlabeled pool set and a test set in a 50% ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_raw, y_raw, test_size=0.5, random_state=42)

# Define our PCA transformer and fit it onto the training set.
pca = PCA(n_components=2, random_state=RANDOM_STATE_SEED)
transformed_X_train = pca.fit_transform(X=X_train)

# Isolate the data we'll need for plotting.
x_component, y_component = transformed_X_train[:, 0], transformed_X_train[:, 1]

# Plot the dimensionality-reduced training set.
plt.figure(figsize=(8.5, 6), dpi=130)
plt.scatter(x=x_component, y=y_component, c=y_train, cmap='viridis', s=50, alpha=8/10)
plt.title('Diabetes classes after PCA transformation')
plt.show()

# Isolate our examples for our labeled dataset.
n_labeled_examples = X_raw.shape[0]
training_indices = np.random.randint(low=0, high=n_labeled_examples + 1, size=4) # size number of training examples are used

X_train = X_raw[training_indices]
y_train = y_raw[training_indices]

# Isolate the non-training examples we'll be querying.
X_pool = np.delete(X_raw, training_indices, axis=0)
y_pool = np.delete(y_raw, training_indices, axis=0)


committee_sizes = []
# Specify the range of committee sizes to test
for i in range(40):
  committee_sizes.append(i)

performance_history_committee = []
best_accuracy = 0.0
best_committee_size = 0

for committee_size in committee_sizes:
    # Define the committee estimators
    committee_estimators = [RandomForestRegressor(n_estimators=10, random_state=seed) for seed in range(committee_size)]

    # Create the committee
    committee = CommitteeRegressor(
        learner_list=[ActiveLearner(estimator=estimator, X_training=X_train, y_training=y_train) for estimator in committee_estimators],
        query_strategy=uncertainty_sampling
    )

    committee_scores = [learner.score(X_test, y_test) for learner in committee.learner_list]
    model_accuracy_committee = np.mean(committee_scores)
    performance_history_committee.append(model_accuracy_committee)

    if model_accuracy_committee > best_accuracy:
        best_accuracy = model_accuracy_committee
        best_committee_size = committee_size

print("------------------------------------------------------------------------------------------------------------------------")

# Plot the performance of different committee sizes
plt.figure(figsize=(8.5, 6), dpi=130)
plt.plot(committee_sizes, performance_history_committee)
plt.xlabel('Committee Size')
plt.ylabel('Accuracy')
plt.title('Performance of Query-by-Committee with Different Committee Sizes')
plt.show()

print(f"Best accuracy: {best_accuracy:.4f}")
print(f"Best committee size: {best_committee_size}")

print("--------------------------------------- Uncertainty Sampling ---------------------------------------------------")

# Specify our core estimator along with its active learning model.
committee_size = best_committee_size  # Ορίζουμε το μεγεθος της επιτροπης με το καλύτερο accuracy

# Ορίζουμε τους εκτιμητές που θα απαρτίζουν την επιτροπή
committee_estimators = [RandomForestRegressor(n_estimators=10, random_state=seed) for seed in range(committee_size)]

# Δημιουργία της επιτροπής
committee = CommitteeRegressor(
    learner_list=[ActiveLearner(estimator=estimator, X_training=X_train, y_training=y_train) for estimator in
                  committee_estimators],
    query_strategy=uncertainty_sampling
)

committee_scores = [learner.score(X_test, y_test) for learner in committee.learner_list]



# Ορισμός μοντέλου με random sampling
N_QUERIES = 30

# query_indices_random = random_sampling(X_pool, N_QUERIES)
query_indices_random = np.random.choice(range(len(X_pool)), size=min(N_QUERIES, len(X_pool)), replace=False)


# Teach the random sampling model the randomly selected records.
X_train_random = np.concatenate((X_train, X_pool[query_indices_random]))
y_train_random = np.concatenate((y_train, y_pool[query_indices_random]))

# Train a random sampling model.
from sklearn.ensemble import RandomForestRegressor

model_random = RandomForestRegressor(random_state=RANDOM_STATE_SEED)

# Train the random model on the updated training set.
model_random.fit(X_train_random, y_train_random)

performance_history_random = [model_random.score(X_test, y_test)]
performance_history_committee = [np.mean(committee_scores)]

# Εκτεούμε την Query-by-Committee για N_QUERIES επαναλήψεις
for index in range(N_QUERIES):
    # Πρόβλεψη των τιμών για κάθε εκτιμητή στην επιτροπή
    committee_predictions = [learner.predict(X_pool) for learner in committee.learner_list]
    
    # Υπολογισμός ασυμφωνίας για κάθε δείγμα
    disagreements = np.std(committee_predictions, axis=0)
    
    # Επιλέγουμε το δείγμα με τη μεγαλύτερη ασυμφωνία για ετικετοποίηση
    query_index_committee = np.argmax(disagreements)
    query_instance_committee = X_pool[query_index_committee].reshape(1, -1)
    
    # Εκπαιδεύουμε το μοντέλο με το επιλεγμένο δείγμα
    committee.teach(query_instance_committee, y_pool[query_index_committee].reshape(1, ))
    
    # Αφαιρούμε το επιλεγμένο δείγμα από το X_pool και το y_pool
    X_pool = np.delete(X_pool, query_index_committee, axis=0)
    y_pool = np.delete(y_pool, query_index_committee)
    
    # Υπολογίζουμε και καταγράφουμε την ακρίβεια του μοντέλου στο test set
    committee_scores = [learner.score(X_test, y_test) for learner in committee.learner_list]
    model_accuracy_committee = np.mean(committee_scores)
    print('Raw accuracy after query {n}: {acc:0.4f}'.format(n=index + 1, acc=model_accuracy_committee))

    performance_history_committee.append(model_accuracy_committee)

print("----------------------------------------- Random Sampling --------------------------------------------------------")

# Allow the random sampling model to query the unlabeled pool set for random points.
for index in range(N_QUERIES):
    query_index_random = np.random.choice(range(len(X_pool)))
    query_instance_random = X_pool[query_index_random]

    X, y = query_instance_random.reshape(1, -1), y_pool[query_index_random].reshape(1, )

    # Remove the queried instance from the unlabeled pool set.
    X_pool, y_pool = np.delete(X_pool, query_index_random, axis=0), np.delete(y_pool, query_index_random)

    # query_indices_random = random_sampling(X_pool, N_QUERIES)
    query_indices_random = np.random.choice(range(len(X_pool)), size=min(N_QUERIES, len(X_pool)), replace=False)

    # Teach the random sampling model the randomly selected records.
    X_train_random = np.concatenate((X_train, X_pool[query_indices_random]))
    y_train_random = np.concatenate((y_train, y_pool[query_indices_random]))

    # Train the random model on the updated training set.
    model_random.fit(X_train_random, y_train_random)

    # Calculate and record the random sampling model's accuracy on the training set.
    accuracy_random = model_random.score(X_test, y_test)
    print('Accuracy after query {n}: {acc:0.4f}'.format(n=index + 1, acc=accuracy_random))

    performance_history_random.append(accuracy_random)


# Plot the performance comparison of uncertainty sampling and random sampling.
print("------------------------------------------------------------------------------------------------------------------------")
query_indices = np.arange(N_QUERIES + 1)
plt.figure(figsize=(8.5, 6), dpi=130)
plt.plot(query_indices, performance_history_random, label='Random Sampling')
plt.plot(query_indices, performance_history_committee, label='Query-by-Committee')
plt.xlabel('Number of Queries')
plt.ylabel('Accuracy')
plt.title('Performance of Uncertainty Sampling')
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from modAL.models import ActiveLearner
from modAL.uncertainty import uncertainty_sampling
from modAL.models import CommitteeRegressor

# Set our RNG seed for reproducibility.
RANDOM_STATE_SEED = 123
np.random.seed(RANDOM_STATE_SEED)


def random_sampling(X, n_samples):
    n_pool = len(X)
    if n_samples > n_pool:
        raise ValueError("Pool set size is smaller than the requested number of samples.")
    query_indices = np.random.choice(range(n_pool), size=n_samples, replace=False)
    return query_indices


# Load the diabetes dataset
diabetes = load_diabetes()
X_raw = diabetes.data
y_raw = diabetes.target

# Split the test set into an unlabeled pool set and a test set in a 50% ratio
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_raw, y_raw, test_size=0.5, random_state=42)

# Define our PCA transformer and fit it onto the training set.
pca = PCA(n_components=2, random_state=RANDOM_STATE_SEED)
transformed_X_train = pca.fit_transform(X=X_train)

# Isolate the data we'll need for plotting.
x_component, y_component = transformed_X_train[:, 0], transformed_X_train[:, 1]

# Plot the dimensionality-reduced training set.
plt.figure(figsize=(8.5, 6), dpi=130)
plt.scatter(x=x_component, y=y_component, c=y_train, cmap='viridis', s=50, alpha=8 / 10)
plt.title('Diabetes classes after PCA transformation')
plt.show()

# Isolate our examples for our labeled dataset.
n_labeled_examples = X_raw.shape[0]
training_indices = np.random.randint(low=0, high=n_labeled_examples + 1,
                                     size=4)  # size number of training examples are used

X_train = X_raw[training_indices]
y_train = y_raw[training_indices]

# Isolate the non-training examples we'll be querying.
X_pool = np.delete(X_raw, training_indices, axis=0)
y_pool = np.delete(y_raw, training_indices, axis=0)

committee_sizes = []
# Specify the range of committee sizes to test
for i in range(40):
    committee_sizes.append(i)

performance_history_committee = []
best_accuracy = 0.0
best_committee_size = 0

for committee_size in committee_sizes:
    # Define the committee estimators
    committee_estimators = [RandomForestRegressor(n_estimators=10, random_state=seed) for seed in range(committee_size)]

    # Create the committee
    committee = CommitteeRegressor(
        learner_list=[ActiveLearner(estimator=estimator, X_training=X_train, y_training=y_train) for estimator in
                      committee_estimators],
        query_strategy=uncertainty_sampling
    )

    committee_scores = [learner.score(X_test, y_test) for learner in committee.learner_list]
    model_accuracy_committee = np.mean(committee_scores)
    performance_history_committee.append(model_accuracy_committee)

    if model_accuracy_committee > best_accuracy:
        best_accuracy = model_accuracy_committee
        best_committee_size = committee_size

print(
    "------------------------------------------------------------------------------------------------------------------------")

# Plot the performance of different committee sizes
plt.figure(figsize=(8.5, 6), dpi=130)
plt.plot(committee_sizes, performance_history_committee)
plt.xlabel('Committee Size')
plt.ylabel('Accuracy')
plt.title('Performance of Query-by-Committee with Different Committee Sizes')
plt.show()

print(f"Best accuracy: {best_accuracy:.4f}")
print(f"Best committee size: {best_committee_size}")

print(
    "--------------------------------------- Uncertainty Sampling ---------------------------------------------------")

# Specify our core estimator along with its active learning model.
committee_size = best_committee_size  # Ορίζουμε το μεγεθος της επιτροπης με το καλύτερο accuracy

# Ορίζουμε τους εκτιμητές που θα απαρτίζουν την επιτροπή
committee_estimators = [RandomForestRegressor(n_estimators=10, random_state=seed) for seed in range(committee_size)]

# Δημιουργία της επιτροπής
committee = CommitteeRegressor(
    learner_list=[ActiveLearner(estimator=estimator, X_training=X_train, y_training=y_train) for estimator in
                  committee_estimators],
    query_strategy=uncertainty_sampling
)

committee_scores = [learner.score(X_test, y_test) for learner in committee.learner_list]

rf = RandomForestRegressor(n_estimators=10, random_state=42)
learner_random = ActiveLearner(estimator=rf, query_strategy=random_sampling, X_training=X_train, y_training=y_train)

N_QUERIES = 30
performance_history_committee = [np.mean(committee_scores)]
performance_history_random = [learner_random.score(X_test, y_test)]

# Εκτεούμε την Query-by-Committee για N_QUERIES επαναλήψεις
for index in range(N_QUERIES):
    # Πρόβλεψη των τιμών για κάθε εκτιμητή στην επιτροπή
    committee_predictions = [learner.predict(X_pool) for learner in committee.learner_list]

    # Υπολογισμός ασυμφωνίας για κάθε δείγμα
    disagreements = np.std(committee_predictions, axis=0)

    # Επιλέγουμε το δείγμα με τη μεγαλύτερη ασυμφωνία για ετικετοποίηση
    query_index_committee = np.argmax(disagreements)
    query_instance_committee = X_pool[query_index_committee].reshape(1, -1)

    # Εκπαιδεύουμε το μοντέλο με το επιλεγμένο δείγμα
    committee.teach(query_instance_committee, y_pool[query_index_committee].reshape(1, ))

    # Αφαιρούμε το επιλεγμένο δείγμα από το X_pool και το y_pool
    X_pool = np.delete(X_pool, query_index_committee, axis=0)
    y_pool = np.delete(y_pool, query_index_committee)

    # Υπολογίζουμε και καταγράφουμε την ακρίβεια του μοντέλου στο test set
    committee_scores = [learner.score(X_test, y_test) for learner in committee.learner_list]
    model_accuracy_committee = np.mean(committee_scores)
    print('Raw accuracy after query {n}: {acc:0.4f}'.format(n=index + 1, acc=model_accuracy_committee))

    performance_history_committee.append(model_accuracy_committee)

print(
    "----------------------------------------- Random Sampling --------------------------------------------------------")

# Allow the random sampling model to query the unlabeled pool set for random points.
for index in range(N_QUERIES):
    query_index_random = np.random.choice(range(len(X_pool)))
    query_instance_random = X_pool[query_index_random]

    # Teach the random sampling model the randomly selected record.
    X, y = query_instance_random.reshape(1, -1), y_pool[query_index_random].reshape(1, )
    learner_random.teach(X=X, y=y)

    # Remove the queried instance from the unlabeled pool set.
    X_pool, y_pool = np.delete(X_pool, query_index_random, axis=0), np.delete(y_pool, query_index_random)

    # Calculate and record the random sampling model's accuracy on the test set.
    model_accuracy_random = learner_random.score(X_test, y_test)
    print('Raw accuracy after query {n}: {acc:0.4f}'.format(n=index + 1, acc=model_accuracy_random))

    performance_history_random.append(model_accuracy_random)

# Plot the performance comparison of uncertainty sampling and random sampling.
print(
    "------------------------------------------------------------------------------------------------------------------------")
query_indices = np.arange(N_QUERIES + 1)
plt.figure(figsize=(8.5, 6), dpi=130)
plt.plot(query_indices, performance_history_committee, label='Query-by-Committee')
plt.plot(query_indices, performance_history_random, label='Random Sampling')
plt.xlabel('Number of Queries')
plt.ylabel('Accuracy')
plt.title('Performance Comparison of Uncertainty Sampling and Random Sampling')
plt.legend(loc='lower right')
plt.show()
```

Αρχικά, η ακρίβεια είναι χαμηλή (περίπου 7-8%), αλλά με τη συνέχιση της επιλογής δειγμάτων, βελτιώνεται και φτάνει περίπου στο 40% μετά από 30 επιλογές. Συνολικά, το Uncertainty Sampling φαίνεται να έχει καλύτερη απόδοση σε αυτά τα δεδομένα, καθώς παρέχει στο μοντέλο πιο ποιοτικά δείγματα για μάθηση και βελτιστοποίηση.
